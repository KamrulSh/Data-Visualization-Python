# pandas multivariate plots

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data_file = "top-five-wine-score-counts.csv"
topFiveData = pd.read_csv(data_file, index_col=0)
topFiveData.head()
```

# which wine most commonly scores an 87?

```python
topFiveData.plot.line(figsize=(16,6))
```

```python
topFiveData.plot.bar(figsize=(16,6))
```

```python
plt.figure(figsize=(16,6))
topFiveData["Pinot Noir"].plot.bar(stacked=True)
```

```python
topFiveData.plot.bar(figsize=(16,6), stacked=True)
```

```python
topFiveData.plot.area(figsize=(16,6), stacked=True)
```

```python
topFiveData.plot.hist(figsize=(16,6), stacked=True)
```

```python
topFiveData[topFiveData<100].plot.pie(figsize=(66,10), subplots=True)
```

```python
topFiveData.plot.pie(figsize=(66,10), subplots=True)
```
