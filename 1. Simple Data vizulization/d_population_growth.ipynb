# Calculating population growth

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
country_data = pd.read_csv("d_population_dataset.csv")
```

```python
country_data
```

# Check if a row is the choosen country

```python
country_data.country == "Bangladesh"
```

# Retrive only the population of Bangladesh

```python
bangladesh = country_data[country_data.country == "Bangladesh"]
```

```python
bangladesh
```

# Ploting the population growth of Bangladesh

```python
plt.plot(bangladesh.year, bangladesh.population / 10**6, "o")
plt.plot(bangladesh.year, bangladesh.population / 10**6)
plt.title("bangladesh population growth")
plt.xlabel("year")
plt.ylabel("population (million)")
plt.show()
```

# Retrive only the population of India

```python
india = country_data[country_data.country == "India"]
```

```python
india
```

# Ploting the population growth of India

```python
plt.plot(india.year, india.population / 10**6, "o")
plt.plot(india.year, india.population / 10**6)
plt.title("IND population growth")
plt.xlabel("year")
plt.ylabel("population (million)")
plt.show()
```

# Comparison of bangladesh and india population

```python
plt.plot(india.year, india.population / 10**6, "o")
plt.plot(india.year, india.population / 10**6)
plt.plot(bangladesh.year, bangladesh.population / 10**6, "s")
plt.plot(bangladesh.year, bangladesh.population / 10**6)
plt.legend(["india plot", "india line", "bangladesh plot", "bangladesh line"])
plt.title("population of BD and IND")
plt.xlabel("year")
plt.ylabel("population (million)")
plt.show()
```

# Population growth of BD in percentage

```python
bd_growth = (bangladesh.population / bangladesh.population.iloc[0]) * 100
```

```python
bd_growth
```

# Population growth of IND in percentage

```python
ind_growth = (india.population / india.population.iloc[0]) * 100
```

```python
ind_growth
```

# Comparison of the population growth of bangladesh and india in percentage

```python
plt.plot(india.year, ind_growth)
plt.plot(bangladesh.year, bd_growth)
plt.legend(["india line", "bangladesh line"])
plt.title("population growth (%) of BD and IND")
plt.xlabel("year")
plt.ylabel("population (first year - 100)")
plt.show()
```
