```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from pandas.plotting import register_matplotlib_converters
```

```python
iris_data = pd.read_csv("j_irisflower_dataset.csv", index_col="Id")
iris_data.head()
```

```python
iris_data.info()
iris_data.describe()
```

```python
plt.figure(figsize=(16,6))
sns.lineplot(x=iris_data["Species"], y=iris_data["Sepal Length (cm)"])
```

```python
plt.figure(figsize=(16,6))
plt.title("Sepal Width (cm) vs Sepal Length (cm)")
sns.lineplot(x=iris_data["Sepal Width (cm)"], y=iris_data["Sepal Length (cm)"])
```

```python
plt.figure(figsize=(16,6))
sns.barplot(data=iris_data)
```

```python
plt.figure(figsize=(16,6))
sns.distplot(a=iris_data["Petal Width (cm)"], kde=False)
```

```python
plt.figure(figsize=(16,6))
sns.distplot(a=iris_data["Petal Length (cm)"])
```

```python
plt.figure(figsize=(16,6))
sns.kdeplot(data=iris_data['Petal Length (cm)'], shade=True)
```

# 2D KDE plots

```python
plt.figure(figsize=(16,6))
sns.jointplot(x=iris_data["Petal Width (cm)"], y=iris_data["Petal Length (cm)"], kind="kde")
# Sepal Length (cm) Sepal Width (cm) Petal Length (cm) Petal Width (cm)
```

```python
plt.figure(figsize=(16,6))
sns.jointplot(x=iris_data["Sepal Length (cm)"], y=iris_data["Petal Length (cm)"])
```

```python
plt.figure(figsize=(16,6))
sns.boxenplot(data=iris_data)
```

# Color-coded plots

```python
iris_setosa = pd.read_csv("j_iris_setosa.csv", index_col="Id")
iris_virgin = pd.read_csv("j_iris_virginica.csv", index_col="Id")
iris_versic = pd.read_csv("j_iris_versicolor.csv", index_col="Id")
iris_setosa.tail()
```

```python
plt.figure(figsize=(16,6))
sns.distplot(a=iris_setosa["Petal Length (cm)"], label="iris setosa", kde=False)
sns.distplot(a=iris_virgin["Petal Length (cm)"], label="iris virginica", kde=False)
sns.distplot(a=iris_versic["Petal Length (cm)"], label="iris versicolor", kde=False)
plt.title("Histogram of Petal Lengths, by Species")
plt.legend()
```

```python
plt.figure(figsize=(16,6))
sns.distplot(a=iris_setosa["Sepal Length (cm)"], label="iris setosa")
sns.distplot(a=iris_virgin["Sepal Length (cm)"], label="iris virginica")
sns.distplot(a=iris_versic["Sepal Length (cm)"], label="iris versicolor")
plt.title("Histogram of Sepal Lengths, by Species")
plt.legend()
```

```python
plt.figure(figsize=(16,6))
sns.kdeplot(data=iris_setosa["Petal Length (cm)"], label="iris setosa", shade=True)
sns.kdeplot(data=iris_virgin["Petal Length (cm)"], label="iris virginica", shade=True)
sns.kdeplot(data=iris_versic["Petal Length (cm)"], label="iris versicolor", shade=True)
plt.title("Histogram of Petal Lengths, by Species")
plt.legend()
```
