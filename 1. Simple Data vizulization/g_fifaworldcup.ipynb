```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from pandas.plotting import register_matplotlib_converters
```

```python
world_cup = pd.read_csv("g_fifaworldcup_dataset.csv", index_col="Date", parse_dates=True)
world_cup.head()
```

```python
world_cup.describe()
```

```python
register_matplotlib_converters()
plt.figure(figsize=(16,6))
sns.lineplot(data=world_cup)
```

```python
plt.figure(figsize=(16,6))
sns.barplot(data=world_cup)
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(data=world_cup)
```
