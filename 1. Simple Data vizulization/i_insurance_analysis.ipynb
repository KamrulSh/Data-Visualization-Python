```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from pandas.plotting import register_matplotlib_converters
```

```python
insurance_data = pd.read_csv("i_insurance_dataset.csv")
insurance_data.head()
```

```python
insurance_data.describe()
```

```python
plt.figure(figsize=(16,6))
plt.title("Age vs charges Graph")
sns.scatterplot(x=insurance_data["age"], y=insurance_data["charges"])
```

```python
plt.figure(figsize=(16,6))
plt.title("bmi vs charges Graph")
sns.scatterplot(x=insurance_data["bmi"], y=insurance_data["charges"], hue=insurance_data["smoker"])
```

```python
plt.figure(figsize=(16,6))
plt.title("bmi vs charges Graph")
sns.regplot(x=insurance_data["bmi"], y=insurance_data["charges"])
```

```python
sns.lmplot(x="bmi", y="charges", hue="smoker", data=insurance_data)
```

```python
plt.figure(figsize=(16,6))
plt.title("smoker vs charges Graph")
sns.swarmplot(x=insurance_data["smoker"], y=insurance_data["charges"])
```

```python
plt.figure(figsize=(16,6))
plt.title("region vs charges Graph")
sns.swarmplot(x=insurance_data["region"], y=insurance_data["charges"])
```
