```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
plt.style.use('seaborn-whitegrid')
plt.figure()
plt.axes()
plt.show()
```

```python
x = np.linspace(0, 10, 1000)
print(x)
```

```python
plt.figure(figsize=(15,6))
plt.plot(x, np.sin(x), label='sin', color='red')
plt.plot(x, np.cos(x), label='cos', color='green')
plt.plot(x, np.sin(x-1), label='sin(x-1)')
plt.plot(x, np.cos(x-1), label='cos(x-1)')
plt.plot(x, 3+np.cos(x), label='3+cosx', color='violet')
plt.title('sin/cos wave')
plt.xlabel('x')
plt.ylabel('sinx / cosx / sin(x-1) / cos(x-1) / 3+cosx')
plt.legend()
plt.show()
```

```python
plt.plot(x, x+1, linestyle='solid')
plt.plot(x, x+3, linestyle='dashed')
plt.plot(x, x+5, linestyle='dashdot')
plt.plot(x, x+7, linestyle='dotted')
```

```python
plt.plot(x, np.power(x,0.5), linestyle='-')
plt.plot(x, x+3, linestyle='--')
plt.plot(x, x+5, linestyle='-.')
plt.plot(x, x+7, linestyle=':')
```

```python
plt.figure(figsize=(16,6))
plt.plot(x, np.sin(x), color='red')
plt.xlim(12, -2)
plt.ylim(1.5, -1.5)
plt.legend(['sin'])
```

```python
plt.figure(figsize=(16,6))
plt.plot(x, np.sin(x), color='red')
plt.xlim(-2, 12)
plt.ylim(-1.5, 1.5)
plt.legend(['sin'])
```
